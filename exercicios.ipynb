```python
# Questão:
# Você está criando um programa para gerenciar uma lista de contatos. Crie um programa em Python que permita ao usuário adicionar novos contatos, visualizar a lista de contatos e procurar um contato pelo nome.

# Requisitos:

# O programa deve ter um menu principal com as opções: adicionar contato, visualizar contatos, procurar contato e sair.
# O usuário deve ser capaz de adicionar um novo contato, fornecendo o nome e o número de telefone.
# O usuário deve ser capaz de visualizar todos os contatos atualmente na lista.
# O usuário deve ser capaz de procurar um contato pelo nome e exibir as informações desse contato.
# O programa deve continuar em execução até que o usuário escolha sair.
```

```python
def visualizar_contato(visualizar_agenda_contato):

  if len(visualizar_agenda_contato) > 0:

    print("Todos os contatos")

    for visualizar_contato in visualizar_agenda_contato :
      print(visualizar_contato, visualizar_agenda_contato[visualizar_contato])
  else:
    print("Agenda Vazia ")

def adicionar_contato(adicionar_agenda_contato):

  print("Adicionar Contato")

  nome_contato = input("Digite o nome do contato")

  try:
    telefone_contato = int(input("Digite o telefone do contato"))
    if len(str(telefone_contato)) < 7 or len(str(telefone_contato)) > 11:
      raise ValueError("Telefone invalido")
    else:
       adicionar_agenda_contato[nome_contato] = telefone_contato
       print("Contato foi adicionado a agenda")

  except ValueError:
    print("Telefone invalido")
  lista_agenda = adicionar_agenda_contato

  return lista_agenda


def remover_contato(remover_agenda_contato):
  print("Remover Contato")
  try:
    remover_nome = input("Digite o nome do contato")
    del remover_agenda_contato[remover_nome]
    print("Contato foi removido com sucesso")
  except KeyError:
    print("Nome digitado não existe na agenda")

  lista_agenda = remover_agenda_contato

  return lista_agenda


def procurar_contato(procurar_agenda_contato):
  print("Procurar Contato")
  procurar_nome = input("Digite o nome do contato")

  if procurar_nome in procurar_agenda_contato:
    print(f"{procurar_nome} - {procurar_agenda_contato[procurar_nome]}")
  else:
    print("Não existe contato com esse nome")



def mensagem_boas_vindas():
  print("--------------------------------------------------")
  print("                    Agenda Telefonica")
  print("--------------------------------------------------")

def mensagem_enceramento():
  print("Encerrando ...")

lista_agenda = {}



menu_opcao = ("Visualizar todos os Contatos", "Criar Contato",
              "Remover Contato", "Procurar Contato", "Sair " )
loop = True
mensagem_boas_vindas()
while loop:
  print(" Digite uma valor")
  for menu_posicao in range(len(menu_opcao)):
    print(f" {menu_posicao} - {menu_opcao[menu_posicao]}")

  try:
    escolha_opcao_contato = int(input())


    if escolha_opcao_contato < 0 and escolha_opcao_contato > 4:
      print("Valor invalidor")
    else:
      if  escolha_opcao_contato == 0:
        visualizar_contato(lista_agenda)

      elif  escolha_opcao_contato == 1:
       lista_agenda = adicionar_contato(lista_agenda)

      elif  escolha_opcao_contato == 2:
         lista_agenda = remover_contato(lista_agenda)

      elif  escolha_opcao_contato == 3:
        procurar_contato(lista_agenda)

      elif escolha_opcao_contato == 4:
        mensagem_enceramento()
        loop = False

  except ValueError:
    print("Digite um valor valido")
```

```python
def adcionar_despesas_gerenciador(gerencia_adcionar_despesas):
  loop_adcionar_despesa = True
  despesas_para_adcionar = ()
  while loop_adcionar_despesa:
    nome_despesa = input("Digite o nome da Despesa ou Sair para voltar ao menu: ")
    if nome_despesa.lower() == "sair":
      loop_adcionar_despesa = False

    else:
      if nome_despesa in gerencia_adcionar_despesas:
        print("Já existe uma despesa com esse nome")

      else:
        print("Digite uma descrição para a depesa")
        descricao_despesa = input()

        try:
          valor_despesa = float(input("Digite o valor da despesa"))

          if type(valor_despesa) == float:
            despesas_para_adcionar = descricao_despesa, valor_despesa
            gerencia_adcionar_despesas[nome_despesa] = despesas_para_adcionar
            print("Despesa foi Adcionada com Sucesso")
            loop_adcionar_despesa = False

        except ValueError:
          print("Digite apenas numeros")




def visualizar_todas_despesas(despesas_para_visualizar):
  print(" Todas as Despesas ")

  if len(despesas_para_visualizar) > 0 :
    for lista_despesas_visualizar in despesas_para_visualizar:
      print("-----------------------------")
      chave_despesa_procurar = despesas_para_visualizar[lista_despesas_visualizar]
      print(f"nome: {lista_despesas_visualizar}")
      print(f"Descrição: {chave_despesa_procurar[0]}")
      print(f"Valor: {chave_despesa_procurar[1]}")
      print("-----------------------------")
  else:
    print("Lista de Despesas está Vazia")





def resurmo_todas_despesas(despesas_para_resurmo):

  if len(despesas_para_resurmo) > 0:
      maior_despesa = 0
      valor_somar_media = 0
      nome_maior_despesa_resurmo = 0
      nome_menor_despesa_resurmo = 0
      menor_despesa = 1000000
      maior_despesa = 0

      for lista_despesa in  despesas_para_resurmo:

        descricao_despesa_resurmo, valor_despesa_resurmo = despesas_para_resurmo[lista_despesa]

        valor_somar_media += valor_despesa_resurmo

        if valor_despesa_resurmo > maior_despesa:
          nome_maior_despesa_resurmo = lista_despesa
          maior_despesa = valor_despesa_resurmo
        else:

          if valor_despesa_resurmo < menor_despesa:
            nome_menor_despesa_resurmo = lista_despesa
            menor_despesa = valor_despesa_resurmo

      print("-----------------------------")
      print("Media Despesa")

      print(f"Valor Medio: {valor_somar_media/len(despesas_para_resurmo)}")
      print("-----------------------------")

      print("-----------------------------")
      print("Maior Despesa")
      print(f"Nome: {nome_maior_despesa_resurmo}")
      print(f"Valor: {maior_despesa}")
      print("-----------------------------")

      print("-----------------------------")
      print("Menor Despesa")
      print(f"Nome: {nome_menor_despesa_resurmo}")
      print(f"Valor: {menor_despesa}")
      print("-----------------------------")

  else:
    print("Lista Vazia")


def procurar_especifica_despesa(despesas_para_procurar):
  print("Procurar Despesa")
  if len(despesas_para_procurar) > 0:
    nome_procurar_despesa = input("Digite o nome da Despesa")
    if not (nome_procurar_despesa in despesas_para_procurar):
      print("Não existe despesa com esse nome")
    else:
      valor_nome_encontrado_despesa = despesas_para_procurar[nome_procurar_despesa]
      print("-----------------------------")
      print(f"Nome: {nome_procurar_despesa}")
      print(f"Descrição \n{valor_nome_encontrado_despesa[0]}")
      print(f"Valor: {valor_nome_encontrado_despesa[1]}")
      print("-----------------------------")
  else:
    print("Lista de Despesas está vazia")

def mensagem_boas_vindas():
  print("--------------------------------------------------")
  print("                  Gerenciador de Despesas")
  print("--------------------------------------------------")

def mensagem_enceramento():
  print("Encerrando ...")


menu = ("Adicionar Despesa", "Visualisar Despesa",
        "Resurmo das Desapesas", "Procurar Despesas",
        "Sair")

cont = 0
loop_start = True
minimo_indice = len(menu) -1

mensagem_boas_vindas()
print("Escolha uma opção")
todas_despesas = {}


while loop_start:

  if cont > minimo_indice:
    cont = 0
  for menu_lista in menu:
    print(f"{cont} - {menu_lista}")
    cont += 1
  menu_opcao_eescolha = input()
  match menu_opcao_eescolha:
    case "0":
      adcionar_despesas_gerenciador(todas_despesas)
    case "1":
      visualizar_todas_despesas(todas_despesas)
    case "2":
      resurmo_todas_despesas(todas_despesas)
    case "3":
      procurar_especifica_despesa(todas_despesas)
    case "4":
      loop_start = False
    case _:
      print("Valor invalido")
      print("Por favor, tente novamente")

mensagem_enceramento()
```

```python
# Exercício: Calculadora Simples

# Crie uma calculadora simples em Python que pode realizar operações de adição,
# subtração, multiplicação e divisão. A calculadora deve ser capaz de receber dois
# números e uma operação do usuário, realizar o cálculo e exibir o resultado.

# Requisitos:

# A calculadora deve continuar pedindo por operações até que o usuário escolha sair.
# Trate erros de entrada, como tentativas de dividir por zero ou entrada inválida.
# Implemente funções separadas para cada operação
#   (adição, subtração, multiplicação, divisão).
# Utilize uma estrutura de repetição para manter
# o programa em execução até que o usuário escolha sair.
```

```python
def conta_adicao(valor_adicao_1, valor_adicao_2):
  return valor_adicao_1 + valor_adicao_2

def conta_subtracao(valor_subtracao_1, valor_subtracao_2):
  return valor_subtracao_1 - valor_subtracao_2

def conta_vezes(valor_vezes_1, valor_vezes_2):
  return valor_vezes_1 * valor_vezes_2

def conta_divisao(valor_divisao_1, valor_divisao_2):
    return valor_divisao_1 / valor_divisao_2

loop_calculadora_simples = True


menu_calculadora_simples = ["0 - Adição", "1 - Subtração", "2 - Multiplicação", "3 - Divisão", "4 - Sair"]

while loop_calculadora_simples:
  print("Digite um valor para fazer a operação")
  for menu_lista_calculadora in menu_calculadora_simples:
    print(menu_lista_calculadora)

  try:
    menu_digitado_calculadora = int(input())

  except ValueError:
    print("Opção invalida")

  if menu_digitado_calculadora >= 0 and menu_digitado_calculadora < 4:

    try:
      valor_digitado_calculadora_1 = int(input("Digite um numero: "))
      valor_digitado_calculadora_2 = int(input("Digite outro numero: "))

      match menu_digitado_calculadora:
        case 0:
          valor_operacao_final = conta_adicao(valor_digitado_calculadora_1, valor_digitado_calculadora_2)
          print(f"{valor_digitado_calculadora_1} + {valor_digitado_calculadora_2} = ", valor_operacao_final)
        case 1:
          valor_operacao_final = conta_subtracao(valor_digitado_calculadora_1, valor_digitado_calculadora_2)
          print(f"{valor_digitado_calculadora_1} - {valor_digitado_calculadora_2} = ", valor_operacao_final)
        case 2:
          valor_operacao_final = conta_vezes(valor_digitado_calculadora_1, valor_digitado_calculadora_2)
          print(f"{valor_digitado_calculadora_1} * {valor_digitado_calculadora_2} = ", valor_operacao_final)
        case 3:
          if valor_digitado_calculadora_1 == 0 or valor_digitado_calculadora_2 == 0:
            print("Um numero não pode ser dividido por zero")
          else:
            valor_operacao_final = conta_divisao(valor_digitado_calculadora_1, valor_digitado_calculadora_2)
            print(f"{valor_digitado_calculadora_1} / {valor_digitado_calculadora_2} = ", valor_operacao_final)

        case _:
          print("Opção invalida")

    except ValueError:
      print("Esse campo aceita apenas valores númericos ")

  if menu_digitado_calculadora == 4:
    loop_calculadora_simples = False
  else:
      print("Valor invalido")
```

```python
# Crie uma matriz 3x3 chamada matriz_a com números inteiros aleatórios entre 1 e 10.
# Crie uma matriz 3x3 chamada matriz_b com números inteiros aleatórios entre 1 e 10.
# Calcule e imprima a multiplicação de elemento a elemento (matriz_a * matriz_b) das duas matrizes.
# Calcule e imprima o determinante da matriz matriz_a.
# Calcule e imprima a matriz inversa da matriz_b.
```

```python
import numpy as np

matriz_a = np.random.randint(1, 10, size=(3, 3))
matriz_b = np.random.randint(1, 10, size=(3, 3))

nova_matriz = np.linalg.det(matriz_a)

print(f"A multiplicação elemento a elemento é \n {matriz_a * matriz_b}")
print(f"A determinante da matriz_a é \n{nova_matriz:.2f}")
```

```python
# Gerar duas matrizes aleatórias (matriz_a e matriz_b).
# Calcular a multiplicação das matrizes.
# Calcular a transposta da matriz matriz_a.
# Calcular a inversa da matriz matriz_b.
```

```python
import numpy as np
nova_matriz_a = np.random.randint(1,10, size=(5,5))
nova_matriz_b = np.random.randint(1,10, size=(5,5))
nova_matrix_multiplicada = np.dot(nova_matriz_a, nova_matriz_b)
nova_matriz_a_transposto = np.transpose(nova_matriz_a)
inverso_nova_matriz_b = np.linalg.inv(nova_matriz_b)
print(f"Multiplicação das matrizes: \n {nova_matrix_multiplicada}")
print(f"O Trnsposto da matriz A é: \n {nova_matriz_a_transposto}")
print(f"O Inverso da matriz B é: \n {inverso_nova_matriz_b}")
```

```python
import pandas as pd
dados_vendas = {
    'Regiao': ['Norte', 'Sul', 'Leste', 'Oeste', 'Central'],
    'Vendas_Janeiro': [500, 700, 900, 600, 800],
    'Vendas_Fevereiro': [600, 750, 850, 700, 900],
    'Vendas_Marco': [550, 800, 950, 650, 1000]
}

# Criando o data Frame
data_frame_vendas = pd.DataFrame(dados_vendas)

data_frame_vendas["Total_Vendas_região"] = data_frame_vendas[['Vendas_Janeiro', 'Vendas_Fevereiro', 'Vendas_Marco']].sum(axis=1)
data_frame_vendas
```

```python
# Calcule o total de vendas (Preço Unitário * Quantidade Vendida)
# para cada produto e adicione essa informação como uma nova
# coluna chamada 'Total_Vendas'.
# Identifique qual é o produto mais vendido (maior quantidade vendida).
# Calcule a média do preço unitário dos produtos em cada categoria.
```

```python
dados_produtos = {
    'Produto': ['A', 'B', 'C', 'D', 'E'],
    'Preco_Unitario': [10.0, 15.5, 20.0, 8.5, 12.0],
    'Quantidade_Vendida': [100, 75, 50, 120, 90],
    'Categoria': ['Eletrônico', 'Moda', 'Eletrônico', 'Alimentos', 'Moda']
}

# Criando o data frame

data_frame_produtos = pd.DataFrame(dados_produtos)

# multiplicando o preço unitario pela quantidade
data_frame_produtos["Total_vendas"] = data_frame_produtos["Preco_Unitario"] * data_frame_produtos["Quantidade_Vendida"]

# Verificando o produto mais vendido
quantidade_mais_vendido = data_frame_produtos['Quantidade_Vendida'].max()

# Agrupando o dataframe e passando a media criando uma nova coluna
data_frame_produtos['Media_Por_Categoria'] = data_frame_produtos.groupby('Categoria')['Preco_Unitario'].transform('mean')


print("Data frame atualizado")
data_frame_produtos.head()

print("Produto mais vendido ")
data_frame_produtos.loc[data_frame_produtos["Quantidade_Vendida"] == quantidade_mais_vendido]
```

```python
import pandas as pd
dados_despesas = {
    'Mês': ['Janeiro', 'Fevereiro', 'Março', 'Abril', 'Maio'],
    'Aluguel': [1200, 1200, 1200, 1200, 1200],
    'Alimentação': [500, 550, 600, 480, 520],
    'Transporte': [150, 180, 200, 160, 170],
    'Educação': [200, 200, 250, 180, 220]
}

# Criando o data frame
df_despesas = pd.DataFrame(dados_despesas)
# Criando a coluna Gastos_total e passando para ela o a soma linha a linha
df_despesas["Gastos_total"] = df_despesas[["Aluguel", "Alimentação", "Transporte", "Educação"]].sum(axis=1)
# Criando a coluna Media_total e passando para ela o a media linha a linha
df_despesas["Media_total"] = df_despesas[["Aluguel", "Alimentação", "Transporte", "Educação"]].mean(axis=1)
print("Més com o maior gasto: ")
df_despesas[df_despesas["Gastos_total"] == df_despesas["Gastos_total"].max()]
df_despesas
```

```python
# Crie um DataFrame chamado df_funcionarios com base nos dados fornecidos.
# Adicione uma nova coluna chamada 'Salario_Ajustado',
# que representa o salário acrescido de 10% para cada funcionário.
# Agrupe os funcionários por departamento e calcule a
# Amédia salarial em cada departamento.


# Calcule a média de idade dos funcionários.
# Encontre o funcionário mais jovem e o mais velho.
# Imprima o DataFrame final.
```

```python
import pandas as pd
dados_funcionarios = {
    'Nome': ['João', 'Maria', 'Carlos', 'Ana', 'Paula'],
    'Idade': [28, 35, 22, 40, 30],
    'Cargo': ['Analista', 'Gerente', 'Desenvolvedor', 'Analista', 'Coordenador'],
    'Salario': [5000, 8000, 3500, 5500, 7000],
    'Departamento': ['TI', 'RH', 'TI', 'Financeiro', 'TI']
}

df_funcionarios = pd.DataFrame(dados_funcionarios)
# Adicionando a coluna preço ajustado
df_funcionarios["Salario_Ajustado"] = df_funcionarios['Salario'] * 1.10
# Criando a coluna media_salarial_departamento r passandop a media por departamento
df_funcionarios["Media_Salarial_Departamento"] = df_funcionarios.groupby("Departamento")["Salario"].transform('mean')
# Informações inportante sobre o Data Frame
media_idade_funcionario = df_funcionarios['Idade'].mean()
minina_idade_funcionario = df_funcionarios['Idade'].min()
maximaa_idade_funcionario = df_funcionarios['Idade'].max()

# Visualizar Data Frame
print(f"A media da idade do funcionarios  é {media_idade_funcionario} ")
print(f"A idade do funcionario mais jovem é {minina_idade_funcionario}")
print(f"A idade do funcionario mais velho é {maximaa_idade_funcionario}")
print(df_funcionarios)
```
